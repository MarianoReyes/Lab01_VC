```python
import numpy as np
import matplotlib.pyplot as plt

original_image = np.zeros((200, 200), dtype=np.uint8)

original_image[30:170, 30:170] = 255
original_image[0:130, 70:130] = 0

```

```python
plt.imshow(original_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')  
plt.show()
```

```python
from skimage.morphology import  dilation, erosion, opening, closing 
from skimage.morphology import square, disk, ball, cube, diamond, ellipse, footprint_from_sequence, octagon, octahedron, rectangle, star 
```

```python
selem = square(25)
slimfit_U= erosion(original_image, selem)

plt.imshow(slimfit_U, cmap='gray')
plt.title('Inciso A')
plt.axis('off')  
plt.show()
```

# El porqué del inciso A


| Elemento estructurante | Operacion morfológica |
| ----- |---- |
| Cuadrado | Erosión |


- La elección de cuadrado es porque la operación a realizar se deseaba hacer con la misma mágnitud de manera verticual y horizontal. El tamaño se fue ajustando a prueba y error. 
- Se realizó una erosión porque se deseaba crear el efecto de disminución de contorno. 


```python
selem = rectangle(125, 20)  # This is an example; you may need to adjust the size
dos_cuadritos =erosion(original_image, selem)

plt.imshow(dos_cuadritos, cmap='gray')
plt.title('Inciso B')
plt.axis('off')  
plt.show()
```

# El porqué del inciso B


| Elemento estructurante | Operacion morfológica |
| ----- |---- |
| Rectangulo | Erosión |


- La elección de un rectangulo es porque se quería un enfoque en contorno verticales y horizontales pero con distintos efectos en cada uno, más que nada para mejorar el ajuste del tamño de las dimensiones del resultados. El tamaño se fue ajustando a prueba y error. 
- Se realizó una erosión porque se deseaba crear el efecto de disminución de contorno. 

```python
selem = ellipse(10, 40)
dos_frijoles = opening(original_image, selem)
plt.imshow(dos_frijoles, cmap='gray')
plt.title('Inciso C')
plt.axis('off')  
plt.show()
```

# El porqué del inciso C


| Elemento estructurante | Operacion morfológica |
| ----- |---- |
| Elipse | Apertura |


- La elección de una elipse es porque es una versión generaliza de un círculo con una definición geométrica lo suficientemente abierta como para alterarla en cuanto a su efecto en ancho y alto, a diferencia de un disco que tiene el mismo tamaño en todas direcciones. En otras palabras, me permitió eliminar la sección horizontal de la U y conservar las verticales a la vez. 
- Se realizó una apertura (que es una erosión tras una dilatación) sirvió para suavizar los bordes de la U (apoyada de la forma natural de una elipse) y la eliminación de los bordes estrechos (sobre todo para la parte inferior de la U).  

```python
selem = disk(10) 
u_ponchada = dilation(original_image, selem)
plt.imshow(u_ponchada, cmap='gray')
plt.title('Inciso D')
plt.axis('off')  
plt.show()
```

# El porqué del inciso D


| Elemento estructurante | Operacion morfológica |
| ----- |---- |
| Disco | Dilatación |


- La elección de un disco es porque su efecto no esta enfocado en bordes horizonatales y verticales sino  en hacerlo todo de manera uniforme.  
- Se realizó una dilatación para aumentar el tamaño del contorno (que en este caso es la U en si) y crear ese efecto de suavizamiento. 

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original')

axes[1].imshow(slimfit_U, cmap='gray')
axes[1].set_title('Inciso A')

axes[2].imshow(dos_cuadritos, cmap='gray')
axes[2].set_title('Inciso B')

axes[3].imshow(dos_frijoles, cmap='gray')
axes[3].set_title('Inciso C')

axes[4].imshow(u_ponchada, cmap='gray')
axes[4].set_title('Inciso D')
for ax in axes:
    ax.axis('off')
plt.tight_layout()
```
